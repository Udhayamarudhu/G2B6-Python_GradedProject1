<font size=4>**4.Convert the columns "Value", "Wage", "Release Clause" to float datatype
after getting rid of currency symbol and suffix.**</font>

```python
df['Value'] = df['Value'].str.replace(r'[€MK]', '', regex=True)
df['Wage'] = df['Wage'].str.replace(r'[€MK]', '', regex=True)
df['Release Clause'] = df['Release Clause'].str.replace(r'[€MK]', '', regex=True)
df['Value'] = pd.to_numeric(df['Value'])
df['Wage'] = pd.to_numeric(df['Wage'])
df['Release Clause'] = pd.to_numeric(df['Release Clause'])
df['Value'] *= 1_000_000
df['Wage'] *= 1_000 
df['Release Clause']*=1_000_000

```

<font size=4>**5. Convert the column "Joined" into integer data type with keeping only the
year**</font>

```python
df['Joined'] = pd.to_datetime(df['Joined'])
df['Joined'] = df['Joined'].dt.year
df['Joined'] = df['Joined'].fillna(0).astype(int)
df['Joined']
```

<font size=4>**6. Convert the column "Contract Valid Until" to pandas datetime type.**</font>

```python
df['Contract Valid Until']= pd.to_datetime(df['Contract Valid Until'],errors='coerce')
df['Contract Valid Until']
```

<font size=4>**7. The column 'Height' is in inches with a quotation mark, Convert to float
with decimal points.**</font>

```python
df['Height'] = df['Height'].str.replace("'", '.').astype(float)

print(df['Height'])
```

<font size=4>**8.The column "Weight" has the suffix as lbs, remove the suffix and convert
to float.**</font>

```python
df['Weight']=df['Weight'].str.replace('lbs','').astype(float)
df['Weight']
```
